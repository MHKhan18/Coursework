### Problem 1 [10 pts]
- Implement a function `dot_product` to calculate the dot product of two 1 by n arrays. [6 pts] 
```
a = array([1, 2, 3, 4])
b = array([2, 3, 4, 5])
dot_product(a,b) = 40
```

````python
def dot_product(lst1 , lst2):
    res = 0
    for i in range(len(lst1)):
        res += lst1[i]*lst2[i]
    return res
    
````

- Test your code after generating two random 1 by 10 arrays. [4 pts]

````python
import numpy as np
lst1 = np.random.randint(1 , 10 , (10,))
lst2 = np.random.randint(1 , 10 , (10,))

print(f'lst1 = {lst1}')
print(f'lst2 = {lst2}')


test = dot_product(lst1 , lst2)
expected = np.dot(lst1 , lst2)


print(f'test = {test}')
print(f'expected = {expected}')
assert test == expected
````

### Problem 3 [20 pts]
- Implement a function `mat_product` to calculate the matrix product of two n by n arrays. [8 pts] For example:


$A = \begin{bmatrix} 1 & 3 & 0 \\ 2 & 0 & 2 \\ 1 & 1 & 4 \end{bmatrix}$ and $A\cdot A=\begin{bmatrix} 7 & 3 & 6 \\ 4 & 8 & 8 \\ 7 & 8 & 18 \end{bmatrix}$

    mat_product(A,A)=?

````python
def mat_product(A,B):
    ''' Only works for square matrices'''
    n = len(A)
    res = []
    for i in range(n):
        res.append([0] * n)
    for i in range(n):
        for j in range(n):
            for k in range(n):
                res[i][j] += A[i][k] * B[k][j]
            
    return res    
````

- Inplement a function `mat_transpose` that transposes ${\bf X}$ and use the `mat_product` function to find the matrice of product: $Z={\bf X}^T\cdot{\bf Y}$ [8 pts]

        Z=mat_product(mat_transpose(X),Y)

````python
def mat_transpose(X):
    rows , cols = len(X) , len(X[0])
    res = []
    for i in range(cols):
        res.append([0] * rows)
    for i in range(rows):
        for j in range(cols):
            res[j][i] = X[i][j]
            
    return res

X = np.random.randint(1 , 10 , (4 , 4))
print(f"X \n {X}")
Y = np.random.randint(1 , 10 , (4 , 4))
print(f"Y \n {Y}")

Z = mat_product(mat_transpose(X),Y)
print(f"Test Z: \n {Z}")

expected = np.matmul(X.T , Y)
print(f"expected \n {expected}")
````

### Problem 5 [30 pts]

- a. Load the Gross Domestic Product data for the 190 ranked countries in the file **GDP.csv**. Load **Country.csv** into a DataFrame. [6 pts]

````python
GDP = pd.read_csv('data/GDP.csv' , skip_blank_lines=True)
GDP.dropna(how="all", inplace=True)
GDP = GDP.iloc[2:192 , 1:5]
GDP = GDP.drop(GDP.columns[[1]] , axis = 1)
GDP.columns = ['Ranking' , 'Country' , 'Value']
GDP
````

````python
Country = pd.read_csv('data/Country.csv' , skip_blank_lines=True)
Country
````

- b. Match the data based on the country shortcode. How many of the IDs match? [6 pts]

````python
Country = Country.rename(columns = {'Short Name' : 'Country'})
GDP_Country = GDP.merge(Country, on='Country')
GDP_Country
````

````python
len(GDP_Country.index)
````

- c. Sort the data frame in descending order by GDP rank. What is the 13th country in the resulting data frame? (Note: GDP formats as 123,456,789.) [6 pts]

````python
GDP_Country['Ranking'] = GDP_Country['Ranking'].astype(float)
GDP_Country = GDP_Country.sort_values(by = 'Ranking' , ascending = False)
GDP_Country
````

````python
GDP_Country.loc[len(GDP_Country.index)-13 , 'Country']
````

- d. What is the average GDP for the "High income: OECD" and "High income: nonOECD" group? [6 pts]

````python
income_grp = GDP_Country.groupby(['Income Group'])
is_oecd = income_grp.get_group("High income: OECD")
oecd_gdps = is_oecd.loc[: , 'Value'].str.replace(',', '').astype(float)
val1 = oecd_gdps.mean()
print(f'Mean GDP of OECD countries: {val1}')
````

````python
non_oecd = income_grp.get_group("High income: nonOECD")
non_oecd_gdps = non_oecd.loc[: , 'Value'].str.replace(',', '').astype(float)
val2 = non_oecd_gdps.mean()
print(f'Mean GDP of Non-OECD countries: {val2}')
````

- e. What is the average GDP for different income groups and regions? [6 pts]

````python
income_grp_gdp = GDP_Country.loc[: , ['Income Group' , 'Value']]
income_grp_gdp['GDP'] = income_grp_gdp['Value'].apply(lambda x : float(x.replace(',', '')))
income_grp_gdp.groupby(['Income Group']).mean()
````

````python
region_gdp = GDP_Country.loc[: , ['Region' , 'Value']]
region_gdp['GDP'] = region_gdp['Value'].apply(lambda x : float(x.replace(',', '')))
region_gdp.groupby(['Region']).mean()
````
